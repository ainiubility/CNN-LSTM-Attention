```python
g = data.groupby('å¸‚åŒº')
df_region = g.count()['å°åŒº']
region = df_region.index.tolist()
count = df_region.values.tolist()
df_pies = df_region.reset_index()
df_pies
```

```python
pie_chart = pygal.Pie(
    inner_radius=.4,  # è§’åº¦
    width=1300,   #å®½åº¦
    height=800,   #é«˜åº¦
    print_values=True,
    print_labels=True,
    print_values_position='top',
    print_labels_position='bottom',
    x_title='åŒ—äº¬å„åŸåŒº',
    y_title='å°åŒº',
    legend_at_bottom=True,       #æ˜¯å¦æ˜¾ç¤ºå›¾ä¾‹
    style=DefaultStyle)
pie_chart.title = 'åŒ—äº¬å„åŸåŒºäºŒæ‰‹æˆ¿å°åŒºåˆ†å¸ƒ'
for i,j in zip(df_pies['å¸‚åŒº'].values.tolist(),df_pies['å°åŒº'].values.tolist()):
    pie_chart.add(
        i,[{'value': int(j), 'label': i}],rounded_bars=10)
HTML(base_html.format(rendered_chart=pie_chart.render(is_unicode=True)))#å›¾ç‰‡æ¸²æŸ“
```
